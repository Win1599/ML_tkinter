```python
#Machine Learning Module in Python(That is externally available and we can use it)
# 1. numpy(Numerical Python)(used for complex calculations)
#  install= pip install numpy
# Example multiplication of Matrix

import numpy as np

#Creating two arrays of rank 2

x=np.array([[1,2],[3,4]])
y=np.array([[5,6],[7,8]])


#Creating two arrays of rank 2

v=np.array([[9,10]])

w=np.array([[11,12]])

# Inner Product of vectors

print(np.dot(v,w))  #single number

#Matrix and Vector Product

print(np.dot(x,v))  # 2 nos


# Matrix and Matrix Product

print(np.dot(x,y)) # 2 by 2 matrix
```
